In this assignment, you are required to implement Depth First Search from scratch.
Your logic is expected to visit the node **before** visiting its neighbors.

### General notes:
* Make sure you tagged the cell that contain your solution with "solution" tag
* Make sure this cell result is independent of any other cells. In other words, it will work even if it was the only cell in the notebook.

```python
def solution():
    def dfs(v, graph):
        """
        v is an integer representing a node (vertex)
        graph is a dictionary where
            - keys are integers representing nodes
            - values are lists, each of them representing the neighbors of v
        """
        res = []

        # write your code here

        def visit(node, visited):
            if node in visited:
                return
            visited.add(node)
            res.append(node)
            neighbors = graph[node]
            for neighbor in neighbors:
                visit(neighbor, visited)

        visited = set()
        visit(v, visited)
        #######################

        """
        return a list of integers in order of visiting each node
        """
        # don't change the return
        return res

    # don't change the return
    return dfs
```

Use the cell below to test your solution before submission.

```python
dfs = solution()
graph = {
        0: [1, 3],
        1: [4],
        2: [4, 5],
        3: [1],
        4: [3],
        5: [5],
    }
true_result = [2, 4, 3, 1, 5]
res = dfs(v=2, graph=graph)
assert true_result == res
print("Test passed successfully!")
```
