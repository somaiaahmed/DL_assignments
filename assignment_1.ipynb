## Assignment 1

In 3D space you have uniformly distributed random points (x<sub>i</sub>,y<sub>i</sub>,z<sub>i</sub>)
You are required to place a sphere with known radius r in 3D space so that the average euclidean distance between the sphere and the points is least.

Use the cell below to define a function that generates 3D data points randomly
in a reproducble way, to test your code before submission.
It is not graded.

```python
from random import Random

def generate_num(N=1000):
  SEED = 5
  random_num = Random(x=SEED)
  return(
      [random_num.uniform(a=1,b=2) for _ in range(N)],
      [random_num.uniform(a=1,b=2) for _ in range(N)],
      [random_num.uniform(a=1,b=2) for _ in range(N)]
  )
data_x, data_y, data_z = generate_num()
     
```

```python
def solution(data_x, data_y, data_z):
    def loss(x_p, y_p, z_p):
        from math import sqrt
        # Implement loss function
        loss = (1/len(data_x))*sum([sqrt((x_i - x_p)**2 + (y_i - y_p)**2 + (z_i - z_p)**2)
                                     for x_i, y_i, z_i in zip(data_x, data_y, data_z)])
        return loss

    # Initialization
    x_p, y_p, z_p = 1, 1, 1
    EPOCHS = 3000
    DELTA = 0.01
    H = 0.001
    epoch_losses = []

    for _ in range(EPOCHS):
        epoch_losses.append(loss(x_p, y_p, z_p))

        # Calculate gradients using numerical differentiation
        grad_x = (loss(x_p + H, y_p, z_p) - loss(x_p, y_p, z_p)) / H
        grad_y = (loss(x_p, y_p + H, z_p) - loss(x_p, y_p, z_p)) / H
        grad_z = (loss(x_p, y_p, z_p + H) - loss(x_p, y_p, z_p)) / H

        # Update sphere position using gradient descent
        x_p -= DELTA * grad_x
        y_p -= DELTA * grad_y
        z_p -= DELTA * grad_z

    return (x_p, y_p, z_p)
```
